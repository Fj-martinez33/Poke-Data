```python
#Antes de nada, como queremos que el algoritmo genere un equipo por nosotros, vamos a guardar un diccionario que relacione el numero de la pokedex con el nombre del pokemon

def CreatePokedex(dataset):
    pkdx_num = dataset["Num"]
    pkdx_name = dataset["Name"]
    nat_pkdx = pd.DataFrame([dataset["Num"], dataset["Name"]])
    nat_pkdx.to_json("../data/processed/pokedex.json")
    return nat_pkdx

CreatePokedex(data)
```

- Eliminación de duplicados

```python
#Vamos a buscar duplicados en la lista y, de haberlos, eliminarlos.

def EraseDuplicates(dataset):

    #Comprobamos la duplicidad.
    if dataset.drop("Num", axis=1).duplicated().sum():
        dataset = dataset.drop_duplicates()
        print ("Duplicates Erased")
    else:
        print("Not Duplicates")

    return dataset

EraseDuplicates(data)

#Como queremos que el algoritmo nos haga un equipo competitivo y los legendarios suelen estar baneados, eliminamos los pokemon legendarios

def EraseLegendaries(dataset):
    dataset.drop(dataset[dataset["Legendary"] == True].index, inplace=True)
    return dataset

#Eliminamos la informacion irrelevante.

irrelevant_list=["Name","Generation","Legendary"] #Insertar aquí las filas irrelevantes. Como hemos eliminado todos los legendarios la fila ya es innecesaria.

def EraseIrrelevantData(dataset, lst):
    dataset.drop(lst, axis=1, inplace=True)
    return dataset

EraseLegendaries(data)
EraseIrrelevantData(data, irrelevant_list)

```
